#Final Data

```python
data = data.rename(columns={"goal_difference": "gd", "opp_goal_difference": "opp_gd", "average_starter_age": "avg_starter_age", "opp_tm_formation": "opp_formation", "tm_formation": "formation", "matchday": "md"})
data
```

```python
result = data.groupby('season')['total_starter_value'].mean()
result = pd.DataFrame(result)
result = result.reset_index()
result
```
